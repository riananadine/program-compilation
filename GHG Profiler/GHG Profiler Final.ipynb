```python
## import
import pandas as pd
import matplotlib.pyplot as plt

gas_data = pd.read_csv('ghg_inventory_processed.csv')
gas_data.head()

## inital input
country = input("Input Country: ")
ghg = input("Input GHG type (CO2 w/o Land Use, GHG w/ Indirect CO2, GHG w/o Land Use, HFCS, CH4, NF3, N2O, PFCS, SF6, Mix): ")

##country filter
country_name = country
countryfiltered_data = gas_data[gas_data['country_or_area'] == country_name]

##ghg type filter
ghg_type = ghg
ghgfiltered_data = countryfiltered_data[countryfiltered_data['Processed_Category'] == ghg_type]

print(ghgfiltered_data)

## selected data
selected_data = ghgfiltered_data[['year','value']]

print(country_name, ghg_type)
print(selected_data)

## Measures of Central Tendency
print("\nMeasures of Central Tendencies")

average_value = ghgfiltered_data['value'].mean()
median_value = ghgfiltered_data['value'].median()
modal_value = ghgfiltered_data['value'].mode()[0]

print("Mean value: ", average_value)
print("Median value: ", median_value)
print("Mode value: ", modal_value)

## Measure of Dispersion
print("\nMeasures of Dispersion")

stdev = ghgfiltered_data['value'].std()
variance = ghgfiltered_data['value'].var()

print("Standard Deviation value: ", stdev)
print("Variance value: ", variance)

## Measure of Position
print("\nMeasures of Position")

data_range = ghgfiltered_data['value'].max()-ghgfiltered_data['value'].min()
q1 = ghgfiltered_data['value'].quantile(0.25)
q3 = ghgfiltered_data['value'].quantile(0.75)
iqr = q3-q1

print("Range value: ", data_range)
print("Quartile 1 value: ", q1)
print("Quartile 2 value: ", q3)
print("IQR value: ", iqr)

## Line graph plot
plt.plot(ghgfiltered_data['year'],ghgfiltered_data['value'],marker = 'o')

plt.title(f'{ghg} Gas Emission Over the Years in {country}')

plt.xlabel('Year')
plt.ylabel('Values - in KT')

plt.grid(True)

plt.show

## Bar graph plot
plt.bar(ghgfiltered_data['year'],ghgfiltered_data['value'],color = 'maroon',width = 0.4)

plt.title(f'{ghg} Gas Emission Over the Years in {country}')

plt.xlabel('Years')
plt.ylabel('Values - in KT')

plt.grid(True)

plt.show
```
