```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

energy_data = pd.read_csv('ASEAN Data Set.csv')
energy_data.head()

corr_matrix = energy_data.corr(numeric_only = True)
print(corr_matrix)

GDP_corr = corr_matrix['gdp']
print(GDP_corr)

## heatmap
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.show()

## Biofuel vs GDP
sns.scatterplot(x='biofuel_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='biofuel_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Biofuel Consumption to GDP')
plt.xlabel('Biofuel Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['biofuel_consumption'], energy_data['gdp'],1)
equation1 =  f"y = {b:.4f} + {m:.4f}x"

print(equation1)

## Coal vs GDP
sns.scatterplot(x='coal_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='coal_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Coal Consumption to GDP')
plt.xlabel('Coal Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['coal_consumption'], energy_data['gdp'],1)
equation2 =  f"y = {b:.4f} + {m:.4f}x"

print(equation2)

## Fossil Fuel vs GDP
sns.scatterplot(x='fossil_fuel_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='fossil_fuel_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Fossil Fuel Consumption to GDP')
plt.xlabel('Fossil Fuel Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['fossil_fuel_consumption'], energy_data['gdp'],1)
equation3 =  f"y = {b:.4f} + {m:.4f}x"

print(equation3)

## Gas vs GDP
sns.scatterplot(x='gas_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='gas_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Gas Consumption to GDP')
plt.xlabel('Gas Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['gas_consumption'], energy_data['gdp'],1)
equation4 =  f"y = {b:.4f} + {m:.4f}x"

print(equation4)

## Hydro vs GDP
sns.scatterplot(x='hydro_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='hydro_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Hydro Consumption to GDP')
plt.xlabel('Hydro Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['hydro_consumption'], energy_data['gdp'],1)
equation5 =  f"y = {b:.4f} + {m:.4f}x"

print(equation5)

## Oil vs GDP
sns.scatterplot(x='oil_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='oil_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Oil Consumption to GDP')
plt.xlabel('Oil Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['oil_consumption'], energy_data['gdp'],1)
equation6 =  f"y = {b:.4f} + {m:.4f}x"

print(equation6)

## Wind vs GDP
sns.scatterplot(x='wind_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='wind_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Wind Consumption to GDP')
plt.xlabel('Wind Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['wind_consumption'], energy_data['gdp'],1)
equation7 =  f"y = {b:.4f} + {m:.4f}x"

print(equation7)

## Other vs GDP
sns.scatterplot(x='other_renewable_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='fossil_fuel_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Other Renewable Consumption to GDP')
plt.xlabel('Other Renewable Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['other_renewable_consumption'], energy_data['gdp'],1)
equation8 =  f"y = {b:.4f} + {m:.4f}x"

print(equation8)

## Solar vs GDP
sns.scatterplot(x='solar_consumption', y = 'gdp', data=energy_data)
sns.regplot(x='solar_consumption',y='gdp',data=energy_data, line_kws={'color':'red'})

plt.title('Solar Consumption to GDP')
plt.xlabel('Solar Consumption')
plt.ylabel('GDP')

m,b = np.polyfit(energy_data['solar_consumption'], energy_data['gdp'],1)
equation9 =  f"y = {b:.4f} + {m:.4f}x"

print(equation9)
```
